# AgriSense Mini Demo

This notebook gives a compact walkthrough of the sample dataset, metadata schema, and saved evaluation reports.

```python
from pathlib import Path
import json
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt

root = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
metadata = pd.read_csv(root / 'data' / 'metadata' / 'sample_metadata.csv')
metadata.head()
```

```python
sample = metadata.iloc[0]
image = Image.open(root / 'data' / 'sample_data' / sample['image_filename'])
plt.figure(figsize=(4, 4))
plt.imshow(image)
plt.axis('off')
plt.title(f"{sample['label']} | {sample['sample_id']}")
plt.show()
```

```python
report_dir = root / 'outputs' / 'reports'
for report_name in ['metadata_reports.json', 'image_reports.json', 'fusion_reports.json']:
    report_path = report_dir / report_name
    if report_path.exists():
        print(f'Loaded: {report_name}')
        print(json.dumps(json.loads(report_path.read_text()), indent=2)[:1500], '\n')
    else:
        print(f'Missing: {report_name}')
```
